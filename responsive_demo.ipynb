# Responsive rawcodebutton - Working Notebook

This notebook contains the actual `rawcodebutton` method that produces the JavaScript to toggle code visibility. Copy this cell into your own notebook to use it.

```python
from IPython.display import display, HTML

# Copy this entire class into your IPython/Jupyter environment
class RawCodeButton:
    """Generate JavaScript to toggle code visibility in Jupyter notebooks"""
    
    def whitebackground(self):
        return '''<style>div.cell { background-color: white; }</style>'''
    
    def rawcodebutton(self, markdown=None):
        """Toggle code visibility - works with JupyterLab and classic Notebook"""
        if markdown is None: markdown = 'inline'
        if markdown == 'right': self.whitebackground()
        out = '''
        <script src="https://code.jquery.com/jquery-3.6.0.min.js" defer></script>
        <script defer>
        var isJupyterLab = false;
        var code_show = true;
        
        function detectEnvironment() {
            if (document.querySelector('.jp-CodeCell')) { isJupyterLab = true; }
            else { isJupyterLab = false; }
        }
        
        function code_toggle() {
            detectEnvironment();
            if (isJupyterLab) {
                if (code_show) { $('.jp-CodeCell .jp-Cell-inputWrapper').hide(); }
                else { $('.jp-CodeCell .jp-Cell-inputWrapper').show(); }
            } else {
                if (code_show) { $('div.input').hide(); } else { $('div.input').show(); }
            }
            code_show = !code_show;
        }
        
        function applyFloatingStyles() {
            detectEnvironment();
            let codeDivName, textDivName, promptDivName;
            if (isJupyterLab) {
                promptDivName = 'div.jp-MarkdownCell div.jp-InputPrompt';
                codeDivName = 'div.jp-CodeCell';
                textDivName = 'div.jp-MarkdownCell';
            } else {
                promptDivName = 'div.text_cell div.prompt';
                codeDivName = 'div.code_cell';
                textDivName = 'div.text_cell';
            }
            $(promptDivName).css({'display':'none'});
            $(codeDivName).css({'width':'75%','float':'left'});
            $(textDivName).css({'width':'25%','float':'right'});
        }
        
        if (markdown === 'inline') {
            applyFloatingStyles = function() {
                detectEnvironment();
                let promptDiv, textDiv;
                if (isJupyterLab) {
                    promptDiv = 'div.jp-MarkdownCell div.jp-InputPrompt';
                    textDiv = 'div.jp-MarkdownCell';
                } else {
                    promptDiv = 'div.text_cell div.prompt';
                    textDiv = 'div.text_cell';
                }
                $(promptDiv).css({'display':'none'});
                $(textDiv).css({'padding-left':'15%','padding-right':'15%','background-color':'silver'});
            };
            applyFloatingStyles();
        } else {
            applyFloatingStyles();
        }
        
        document.addEventListener('DOMContentLoaded', function() { code_toggle(); });
        document.addEventListener('keydown', function(event) {
            if (event.ctrlKey && event.key === 'q') {
                event.preventDefault();
                code_toggle();
            }
        });
        </script>
        <form action="javascript:code_toggle()"><input type="submit" value="Toggle Code (Ctrl+Q)"></form>
        <style>
        /* ====== RESPONSIVE STYLES ====== */
        h1 { padding-top:30px; }
        h3 { padding-top:0px; margin-top:0px !important; padding-bottom:16px; }
        @font-face { font-family: "Segoe UI Emoji"; src: local("Arial"); }
        
        /* Medium screens (tablets in landscape) */
        @media screen and (max-width: 1024px) {
            div.code_cell, div.jp-CodeCell { width: 65% !important; float: left; }
            div.text_cell, div.jp-MarkdownCell { width: 35% !important; float: right; }
        }
        
        /* Small tablets / large phones */
        @media screen and (max-width: 768px) {
            div.code_cell, div.jp-CodeCell { 
                width: 100% !important; 
                float: none;
                clear: both;
            }
            div.text_cell, div.jp-MarkdownCell { 
                width: 100% !important; 
                float: none;
                clear: both;
                padding: 10px !important;
            }
        }
        
        /* Mobile phones */
        @media screen and (max-width: 480px) {
            div.code_cell, div.jp-CodeCell { width: 100% !important; }
            div.text_cell, div.jp-MarkdownCell { 
                width: 100% !important; 
                padding: 8px !important;
                font-size: 14px !important;
            }
            form input[type="submit"] {
                font-size: 12px;
                padding: 6px 10px;
            }
        }
        </style>
        <script defer>
        if (typeof observer === 'undefined') {
            const observer = new MutationObserver((mutations) => {
                mutations.forEach((mutation) => {
                    if (mutation.addedNodes.length) { applyFloatingStyles(); }
                });
            });
            observer.observe(document.body, { childList: true, subtree: true });
        }
        </script>
        '''
        return display(HTML(out))

# Create instance to use
rawcode = RawCodeButton()
```

## Now call the method - markdown on the right

Call `rawcode.rawcodebutton(markdown='right')` to put markdown on the right side of plots.

```python
# This generates and runs the JavaScript
rawcode.rawcodebutton(markdown='right')
```

# Responsive rawcodebutton - Working Notebook

This notebook contains the actual `rawcodebutton` method - copy this cell into your project.

```python
from IPython.display import display, HTML

# Your original method with responsive CSS added
class RawCodeButton:
    
    def whitebackground(self):
        return '''<style>div.cell { background-color: white; }</style>'''
    
    def rawcodebutton(self, markdown=None):
        if markdown is None: markdown = 'inline'
        if markdown == 'right': self.whitebackground()
        out = '''
            <script src="https://code.jquery.com/jquery-3.6.0.min.js" defer></script>
            <script defer>
            // Initialize a flag to track the environment (JupyterLab or classic Notebook)
            var isJupyterLab = false;
            
            // Function to detect whether it's JupyterLab or classic Jupyter Notebook
            function detectEnvironment() {
                // JupyterLab typically uses the "jp-CodeCell" class for code cells
                if (document.querySelector('.jp-CodeCell')) {
                    isJupyterLab = true;
                } else {
                    // Classic Notebook uses "input" or "input_area" class
                    isJupyterLab = false;
                }
            }            
            // Toggle function that hides/shows code based on the detected environment
            function code_toggle() {
                detectEnvironment();  // Check the environment each time
            
                if (isJupyterLab) {
                    if (code_show) { $('.jp-CodeCell .jp-Cell-inputWrapper').hide(); }
                    else { $('.jp-CodeCell .jp-Cell-inputWrapper').show(); }
                } else {
                    if (code_show) { $('div.input').hide(); } else { $('div.input').show(); }
                }
                code_show = !code_show;
            }
            '''
        if markdown in ['inline']:
            out += '''
            // Apply floating styles to code and text cells
            function applyFloatingStyles() {
                detectEnvironment();
                let promptDiv,textDiv;
                
                if (isJupyterLab) {
                    promptDiv = 'div.jp-MarkdownCell div.jp-InputPrompt';
                    textDiv   = 'div.jp-MarkdownCell';
                    }
                else {
                    promptDiv = 'div.text_cell div.prompt';
                    textDiv = 'div.text_cell';
                    }
                $(promptDiv).css({'display':'none'});
                $(textDiv).css({'padding-left':'15%','padding-right':'15%','background-color':'silver'});
            }
            applyFloatingStyles();
            '''
        else:
            out += '''
            // Apply floating styles to code and text cells
            function applyFloatingStyles() {
                detectEnvironment();
                let codeDivName,textDivName,promptDivName;
                
                if (isJupyterLab) {
                    promptDivName = 'div.jp-MarkdownCell div.jp-InputPrompt';
                    codeDivName = 'div.jp-CodeCell';
                    textDivName = 'div.jp-MarkdownCell';
                    $('.jp-cell-toolbar').css({'display':'none !important'});
                    }
                else {
                    promptDivName = 'div.text_cell div.prompt';
                    codeDivName = 'div.code_cell';
                    textDivName = 'div.text_cell';
                    }
                $(promptDivName).css({'display':'none'});
                $(codeDivName).css({'width':'75%','float':'left'});
                $(textDivName).css({'width':'25%','float':'right'});
            }
            applyFloatingStyles();
            ''' 
        out += '''
            // Variable to track the current state of code visibility
            code_show = true;
            
            // Run the toggle function on document ready
            document.addEventListener('DOMContentLoaded', function() {
                code_toggle();
            });
            
            // Add the keydown event listener for Ctrl+Q shortcut
            document.addEventListener('keydown', function(event) {
                // Check if Ctrl+Q is pressed
                if (event.ctrlKey && event.key === 'q') {
                    event.preventDefault();  // Prevent the default browser behavior
                    code_toggle();  // Toggle code visibility
                }
            });
            </script>
            <form action="javascript:code_toggle()"><input type="submit" value="Click here to toggle raw code (Ctrl+Q)"></form>
            <style>
            h1 {
                padding-top:30px;
                }
            h3 {
                padding-top:0px;
                margin-top:0px !important;
                padding-bottom:16px;
                }
            @font-face {
                font-family: "Segoe UI Emoji";
                src: local("Arial"); /* Replace with a font of your choice or use any generic one */
                }
            
            /* ====== NEW: RESPONSIVE STYLES ====== */
            
            /* Medium screens (tablets in landscape) */
            @media screen and (max-width: 1024px) {
                div.code_cell, div.jp-CodeCell { width: 65% !important; float: left; }
                div.text_cell, div.jp-MarkdownCell { width: 35% !important; float: right; }
            }
            
            /* Small tablets / large phones */
            @media screen and (max-width: 768px) {
                div.code_cell, div.jp-CodeCell { 
                    width: 100% !important; 
                    float: none;
                    clear: both;
                    overflow: visible;
                }
                div.text_cell, div.jp-MarkdownCell { 
                    width: 100% !important; 
                    float: none;
                    clear: both;
                    padding: 10px !important;
                }
            }
            
            /* Mobile phones */
            @media screen and (max-width: 480px) {
                div.code_cell, div.jp-CodeCell { width: 100% !important; overflow: visible; }
                div.text_cell, div.jp-MarkdownCell { 
                    width: 100% !important; 
                    padding: 8px !important;
                    font-size: 14px !important;
                }
                form input[type="submit"] {
                    font-size: 12px;
                    padding: 6px 10px;
                }
            }
            </style>
            <script defer>
            // Set up a MutationObserver to monitor for new cells
            if (typeof observer === 'undefined') {
                const observer = new MutationObserver((mutations) => {
                    mutations.forEach((mutation) => {
                        if (mutation.addedNodes.length) {
                            applyFloatingStyles();  // Reapply styles when new nodes are added
                        }
                    });
                });
            
                // Start observing the document body for added child nodes
                observer.observe(document.body, {
                    childList: true,
                    subtree: true,
                });
            }
            </script>
            '''
        return display(HTML(out))

# Create instance to use
rawcode = RawCodeButton()
```

## Call the method - markdown on the right

```python
# Run this to activate the toggle and responsive styles
rawcode.rawcodebutton(markdown='right')
```

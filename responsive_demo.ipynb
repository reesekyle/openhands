# Responsive rawcodebutton Demo

This notebook shows how to make `rawcodebutton` work on smaller screens by adding CSS media queries.

```python
from IPython.display import display, HTML
import textwrap
```

## Original Method (for reference)

The original method uses floats with a fixed 75%/25% split which breaks on mobile.

```python
# Here's the original method for reference
original_code = textwrap.dedent('''
    def rawcodebutton(self, markdown=None):
        if markdown is None: markdown = 'inline'
        if markdown == 'right': self.whitebackground()
        out = '''
            <script src="https://code.jquery.com/jquery-3.6.0.min.js" defer></script>
            <script defer>
            var isJupyterLab = false;
            
            function detectEnvironment() {
                if (document.querySelector('.jp-CodeCell')) { isJupyterLab = true; }
                else { isJupyterLab = false; }
            }            
            function code_toggle() {
                detectEnvironment();
                if (isJupyterLab) {
                    if (code_show) { $('.jp-CodeCell .jp-Cell-inputWrapper').hide(); }
                    else { $('.jp-CodeCell .jp-Cell-inputWrapper').show(); }
                } else {
                    if (code_show) { $('div.input').hide(); } else { $('div.input').show(); }
                }
                code_show = !code_show;
            }
            '''
        if markdown in ['inline']:
            out += '''
            function applyFloatingStyles() {
                detectEnvironment();
                let promptDiv, textDiv;
                if (isJupyterLab) {
                    promptDiv = 'div.jp-MarkdownCell div.jp-InputPrompt';
                    textDiv   = 'div.jp-MarkdownCell';
                } else {
                    promptDiv = 'div.text_cell div.prompt';
                    textDiv = 'div.text_cell';
                }
                $(promptDiv).css({'display':'none'});
                $(textDiv).css({'padding-left':'15%','padding-right':'15%','background-color':'silver'});
            }
            applyFloatingStyles();
            '''
        else:
            out += '''
            function applyFloatingStyles() {
                detectEnvironment();
                let codeDivName, textDivName, promptDivName;
                if (isJupyterLab) {
                    promptDivName = 'div.jp-MarkdownCell div.jp-InputPrompt';
                    codeDivName = 'div.jp-CodeCell';
                    textDivName = 'div.jp-MarkdownCell';
                    $('.jp-cell-toolbar').css({'display':'none !important'});
                } else {
                    promptDivName = 'div.text_cell div.prompt';
                    codeDivName = 'div.code_cell';
                    textDivName = 'div.text_cell';
                }
                $(promptDivName).css({'display':'none'});
                $(codeDivName).css({'width':'75%','float':'left'});
                $(textDivName).css({'width':'25%','float':'right'});
            }
            applyFloatingStyles();
            ''' 
        out += '''
            code_show = true;
            
            document.addEventListener('DOMContentLoaded', function() { code_toggle(); });
            document.addEventListener('keydown', function(event) {
                if (event.ctrlKey && event.key === 'q') {
                    event.preventDefault();
                    code_toggle();
                }
            });
            </script>
            <form action="javascript:code_toggle()"><input type="submit" value="Toggle Code (Ctrl+Q)"></form>
            <style>
            h1 { padding-top:30px; }
            h3 { padding-top:0px; margin-top:0px !important; padding-bottom:16px; }
            @font-face { font-family: "Segoe UI Emoji"; src: local("Arial"); }
            </style>
            <script defer>
            if (typeof observer === 'undefined') {
                const observer = new MutationObserver((mutations) => {
                    mutations.forEach((mutation) => {
                        if (mutation.addedNodes.length) { applyFloatingStyles(); }
                    });
                });
                observer.observe(document.body, { childList: true, subtree: true });
            }
            </script>
            '''
        return display(HTML(out))
''')
display(HTML('<h3>Original Method - Fixed Widths (breaks on mobile)</h3>'))
display(HTML('<pre>' + original_code[:800] + '...</pre>'))
```

## Key Problems with the Original Code

1. **Fixed width percentages**: `width: 75%` and `width: 25%` don't adapt to screen size
2. **No media queries**: No responsive breakpoints for tablets/phones
3. **Float layout**: Falls apart on small screens - no wrapping/stacking
4. **No min-width checks**: Can get crushed on very narrow screens

## Solution: Responsive rawcodebutton Method

Here's the tweaked method with CSS media queries added.

```python
responsive_code = textwrap.dedent('''
    def rawcodebutton(self, markdown=None):
        if markdown is None: markdown = 'inline'
        if markdown == 'right': self.whitebackground()
        out = '''
            <script src="https://code.jquery.com/jquery-3.6.0.min.js" defer></script>
            <script defer>
            var isJupyterLab = false;
            
            function detectEnvironment() {
                if (document.querySelector('.jp-CodeCell')) { isJupyterLab = true; }
                else { isJupyterLab = false; }
            }            
            function code_toggle() {
                detectEnvironment();
                if (isJupyterLab) {
                    if (code_show) { $('.jp-CodeCell .jp-Cell-inputWrapper').hide(); }
                    else { $('.jp-CodeCell .jp-Cell-inputWrapper').show(); }
                } else {
                    if (code_show) { $('div.input').hide(); } else { $('div.input').show(); }
                }
                code_show = !code_show;
            }
            '''
        if markdown in ['inline']:
            out += '''
            function applyFloatingStyles() {
                detectEnvironment();
                let promptDiv, textDiv;
                if (isJupyterLab) {
                    promptDiv = 'div.jp-MarkdownCell div.jp-InputPrompt';
                    textDiv   = 'div.jp-MarkdownCell';
                } else {
                    promptDiv = 'div.text_cell div.prompt';
                    textDiv = 'div.text_cell';
                }
                $(promptDiv).css({'display':'none'});
                $(textDiv).css({'padding-left':'15%','padding-right':'15%','background-color':'silver'});
            }
            applyFloatingStyles();
            '''
        else:
            out += '''
            function applyFloatingStyles() {
                detectEnvironment();
                let codeDivName, textDivName, promptDivName;
                if (isJupyterLab) {
                    promptDivName = 'div.jp-MarkdownCell div.jp-InputPrompt';
                    codeDivName = 'div.jp-CodeCell';
                    textDivName = 'div.jp-MarkdownCell';
                    $('.jp-cell-toolbar').css({'display':'none !important'});
                } else {
                    promptDivName = 'div.text_cell div.prompt';
                    codeDivName = 'div.code_cell';
                    textDivName = 'div.text_cell';
                }
                $(promptDivName).css({'display':'none'});
                $(codeDivName).css({'width':'75%','float':'left'});
                $(textDivName).css({'width':'25%','float':'right'});
            }
            applyFloatingStyles();
            ''' 
        out += '''
            code_show = true;
            
            document.addEventListener('DOMContentLoaded', function() { code_toggle(); });
            document.addEventListener('keydown', function(event) {
                if (event.ctrlKey && event.key === 'q') {
                    event.preventDefault();
                    code_toggle();
                }
            });
            </script>
            <form action="javascript:code_toggle()"><input type="submit" value="Toggle Code (Ctrl+Q)"></form>
            <style>
            /* ====== RESPONSIVE STYLES ====== */
            h1 { padding-top:30px; }
            h3 { padding-top:0px; margin-top:0px !important; padding-bottom:16px; }
            @font-face { font-family: "Segoe UI Emoji"; src: local("Arial"); }
            
            /* Medium screens (tablets in landscape) */
            @media screen and (max-width: 1024px) {
                div.code_cell, div.jp-CodeCell { width: 65% !important; float: left; }
                div.text_cell, div.jp-MarkdownCell { width: 35% !important; float: right; }
            }
            
            /* Small tablets / large phones */
            @media screen and (max-width: 768px) {
                div.code_cell, div.jp-CodeCell { 
                    width: 100% !important; 
                    float: none;
                    clear: both;
                }
                div.text_cell, div.jp-MarkdownCell { 
                    width: 100% !important; 
                    float: none;
                    clear: both;
                    padding: 10px !important;
                }
            }
            
            /* Mobile phones */
            @media screen and (max-width: 480px) {
                div.code_cell, div.jp-CodeCell { width: 100% !important; }
                div.text_cell, div.jp-MarkdownCell { 
                    width: 100% !important; 
                    padding: 8px !important;
                    font-size: 14px !important;
                }
                form input[type="submit"] {
                    font-size: 12px;
                    padding: 6px 10px;
                }
            }
            </style>
            <script defer>
            if (typeof observer === 'undefined') {
                const observer = new MutationObserver((mutations) => {
                    mutations.forEach((mutation) => {
                        if (mutation.addedNodes.length) { applyFloatingStyles(); }
                    });
                });
                observer.observe(document.body, { childList: true, subtree: true });
            }
            </script>
            '''
        return display(HTML(out))
''')
# Show the responsive version
display(HTML('<h3>Responsive Method - With Media Queries</h3>'))
display(HTML('<pre>' + responsive_code[:1200] + '</pre>'))
display(HTML('<p><strong>Key changes:</strong> Added three @media query breakpoints at 1024px, 768px, and 480px that stack cells vertically on smaller screens.</p>'))
```

## Summary of Changes

Here's what changed to make it responsive:

| Breakpoint | Width | Behavior |
|-----------|-------|----------|
| > 1024px | 75% / 25% | Original side-by-side |
| 768-1024px | 65% / 35% | Compressed side-by-side |
| 480-768px | 100% / 100% | Stacked vertically |
| < 480px | 100% / 100% | Stacked + smaller fonts |

The key addition is the `<style>` block with three `@media` queries that:
1. **Adjusts** the widths at 1024px (tablet landscape)
2. **Stacks** cells vertically at 768px (portrait tablet/large phone)
3. **Scales down** fonts at 480px (mobile phones)
